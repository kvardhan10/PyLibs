```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/NIFTY_50'
html = pd.read_html(url, header = 0)
df = html[1]
df
```

```python
sectors = df['Sector'].unique()
sectors
```

```python
df.groupby('Sector').count()
```

```python
tickers = list(df.Symbol)
for i in range(0, len(tickers)):
    tickers[i] = tickers[i] + '.NS'
tickers
```

```python
import yfinance as yf
```

```python
data = yf.download(tickers=tickers, period='ytd', interval='1d', group_by='ticker')
```

```python
data
```

```python
data['M&M.NS']
```

```python
from matplotlib import pyplot as plt
```

```python
def plot(symbol):
    plot_df = pd.DataFrame(data[symbol].High)
    plot_df['Date'] = plot_df.index
    plt.fill_between(plot_df.Date, plot_df.High,alpha=0.2)
    
    plt.plot(plot_df.Date, plot_df.High, alpha=1)
    plt.xlabel('Date')
    plt.ylabel('High')
    return plt.show()
```

```python
for ticker in tickers[:10]:
    plot(ticker)
```
