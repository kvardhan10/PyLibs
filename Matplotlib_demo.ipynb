## Simple Line Plot

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.title('Demo for Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show
```

```python
plt.plot([1,2,3],[7,6,9])
plt.show()
```

```python
plt.plot(x,y,color='purple',linestyle='-.',linewidth=2)
# supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted' for linestyle
plt.title('Demo for Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show
```

```python
y2 = 3*x
y2
```

```python
plt.plot(x,y,color='purple',label='line-one',linestyle='-.',linewidth=2)
plt.plot(x,y2,color='cyan',label='line-two',linestyle='-.',linewidth=2)
plt.title('Demo for Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
square = pow(x,2)
cube = pow(x,3)
plt.plot(x,square,color='purple',linestyle='-.',linewidth=2)
plt.plot(x,cube,color='cyan',linestyle='-.',linewidth=2)
plt.title('Demo for Line Plot [cyan:cubes, purple:squares]')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

## Subplots

```python
# subplot() method takes 3 args: x,y,z
# x: no. of rows
# y: no. of cols
# z: index no.    
# SYNTAX: plt.subplot(x,y,z)

plt.subplot(1,2,1) # first row, first col, index 1
plt.plot(x,square,color='r',linestyle='dotted',linewidth=2)
plt.title('Squares subplot on 1 - R1C1')

plt.subplot(1,2,2) # first row, second col, index 2
plt.plot(x,cube,color='b',linestyle='dotted',linewidth=2)
plt.title('Cubes subplot on 2 - R1C2')

plt.show()
```

```python
plt.subplot(2,1,1) # first row, first col, index 1
plt.plot(x,square,color='r',linestyle='dotted',linewidth=2)

plt.subplot(2,1,2) # second row, first col, index 2
plt.plot(x,cube,color='b',linestyle='dotted',linewidth=2)

plt.show()
```

## Bar Plot - Used for categorical data

```python
import random
marks = []
students = ['Alice', 'Bob', 'Chris']
marks = random.sample(range(50,95),3)

plt.subplot(1,2,1)
plt.bar(students,marks,label='line-one',color='r') # bar for vertical
plt.grid(True)
plt.legend()
plt.subplot(1,2,2) #barh for horizontal bar
plt.barh(students,marks,label='line-one',color='g')
plt.legend()
plt.grid(True)
# Add grid, x-label, y-label if needed
# SYNTAX = plt.bar(categorical_values, numerical_vales)

plt.show()
```

```python
a = random.sample(range(1,11),10)
b = random.sample(range(1,11),10)
c = random.sample(range(1,11),10)


plt.subplot(2,2,1)
plt.scatter(a,b,marker='1',c='r') #https://matplotlib.org/stable/api/markers_api.html - diff. markers
plt.grid(True)

plt.subplot(2,2,2)
plt.scatter(b,a)
plt.grid(True)

plt.subplot(2,2,3)
plt.scatter(b,a,marker='8',c='r')
plt.scatter(a,c,marker='s',c='g')
plt.grid(True)

plt.show()

# default marker is blue dots
```

```python
data = [random.randint(1,10) for i in range(0,21)]
print(data)
plt.hist(data)
plt.show()

# so, basically the histogram shows how many times the certain value is in the list
# 1 has appeared 2 times, 2 has appeared 5 times, 3 has appeared 1 time, 4 has appeared 3 times and so on...
```

```python
# each building kinda thingy are called 'bins'
data = [1,3,3,3,3,4,5,5,6,6,6,7,8,9]

plt.subplot(1,2,1)
plt.hist(data,color='g',bins=4)

plt.subplot(1,2,2)
plt.hist(data)

plt.show()
```

```python
fruit = ['apple','mango','orange','pear']
quantity = [random.randint(1,11) for i in range(0,4)]
quantity = [1,2,3,4]

print(quantity)

plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['red','yellow','orange','green'],radius=2,startangle=35,shadow=True,explode=(0.1,0.1,0.1,0.1))
plt.show()
```

```python
days = [1,2,3,4,5]

sleeping = [random.randint(3,10) for i in range(0,5)]
working = [random.randint(3,10) for i in range(0,5)]
playing = [random.randint(3,10) for i in range(0,5)]
eating = [random.randint(3,10) for i in range(0,5)]

plt.stackplot(days, sleeping,working,eating,playing)
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)
plt.plot(t1,f(t1) , 'bo' ,t2,np.cos(2*np.pi*t2) , 'g1' ) 
#bo = blue dot(0), g1 = green tri_down(1) REFER:https://matplotlib.org/stable/api/markers_api.html

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))

plt.show()
```
